# Train mnist with epoch from 10 to 30 with step 10

#   and hiden layer sizes like 128, 256, 512 and two hiden layer with 128, 128 neurones

#   using static learning rate

```python
for n_epoch in (10,20,30):
    for hidden_layer_size in ((128,) ,
                              (256,),
                              (512,),
                              (128,128)):
        params = f'N epoch={n_epoch}_____hidden_layer_size={hidden_layer_size}'
        print ('{:_^120}'.format(params))
        main(n_epochs=n_epoch, hidden_layer_sizes=hidden_layer_size, lr=1e-2)
```

# Using dynamic learning rate

```python
for n_epoch in (10,20,30):
    for hidden_layer_size in ((128,) ,
                              (256,),
                              (512,),
                              (128,128)):
        params = f'N epoch={n_epoch}_____hidden_layer_size={hidden_layer_size}'
        print ('{:_^120}'.format(params))
        main_dynamic_lr(n_epochs=n_epoch, hidden_layer_sizes=hidden_layer_size, lr=1e-1)
```
